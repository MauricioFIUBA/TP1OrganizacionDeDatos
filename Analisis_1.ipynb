```python
df = df.rename(columns={'Pricing, Delivery_Terms_Quote_Appr': 'Pricing, Delivery_Terms_Quote_Approval'})
```

```python
df['Pricing, Delivery_Terms_Quote_Approval'] = df['Pricing, Delivery_Terms_Quote_Approval'].astype('bool')

df['Account_Created_Date'] = df['Account_Created_Date'].astype('datetime64')
df['Opportunity_Created_Date'] = df['Opportunity_Created_Date'].astype('datetime64')
df['Quote_Expiry_Date'] = df['Quote_Expiry_Date'].astype('datetime64')
df['Last_Modified_Date'] = df['Last_Modified_Date'].astype('datetime64')
df['Planned_Delivery_Start_Date'] = df['Planned_Delivery_Start_Date'].astype('datetime64')
df['Planned_Delivery_End_Date'] = df['Planned_Delivery_End_Date'].astype('datetime64')
df['Month'] = df['Month'].astype('datetime64')
df['Actual_Delivery_Date'] = df['Actual_Delivery_Date'].astype('datetime64')
```

```python
region_territorio= df.groupby(['Region','Territory']).agg({'Territory':['count']})
region_territorio
```

```python
level0 = region_territorio.columns.get_level_values(0)
level1 = region_territorio.columns.get_level_values(1)
region_territorio.columns= level0 + '_' + level1
region_territorio =region_territorio.reset_index()
region_territorio
```

```python
region_territorio = region_territorio.dropna()
region_territorio
```

```python
region_territorio = region_territorio[region_territorio['Territory'] != 'None']
region_territorio['Territory_count'] = region_territorio['Territory_count'].astype('int')
region_territorio.info()
```

```python
#region_territorio = region_territorio.sort_values(by='Territory_count', ascending=False)
region_territorio
```

```python
# libraries
import matplotlib.pyplot as plt
from matplotlib import rc

plt.figure(figsize=(10, 20)) 
#region_territorio.Territory[region_territorio.Territory_count.count()].plot(kind='barh', alpha=0.8)
plt.title('Territorios de las distintas regiones', size=22)
plt.xlabel('Cantidad', size=20)
plt.ylabel('Territorios',size=20)

y_pos = np.arange(len(region_territorio['Territory']))
 
# Create horizontal bars
plt.barh(y_pos,region_territorio['Territory_count'].sort_values() )
 
# Create names on the y-axis
plt.yticks(y_pos,region_territorio['Territory'])
 
# Show graphic
plt.show()
```
