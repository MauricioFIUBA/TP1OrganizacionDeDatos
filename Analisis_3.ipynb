```python
df = df.rename(columns={'Pricing, Delivery_Terms_Quote_Appr': 'Pricing, Delivery_Terms_Quote_Approval'})
```

```python
df['Pricing, Delivery_Terms_Quote_Approval'] = df['Pricing, Delivery_Terms_Quote_Approval'].astype('bool')

df['Account_Created_Date'] = df['Account_Created_Date'].astype('datetime64')
df['Opportunity_Created_Date'] = df['Opportunity_Created_Date'].astype('datetime64')
df['Quote_Expiry_Date'] = df['Quote_Expiry_Date'].astype('datetime64')
df['Last_Modified_Date'] = df['Last_Modified_Date'].astype('datetime64')
df['Planned_Delivery_Start_Date'] = df['Planned_Delivery_Start_Date'].astype('datetime64')
df['Planned_Delivery_End_Date'] = df['Planned_Delivery_End_Date'].astype('datetime64')
df['Month'] = df['Month'].astype('datetime64')
df['Actual_Delivery_Date'] = df['Actual_Delivery_Date'].astype('datetime64')
```

```python
# Como estoy haciendo muy seguido el tipo de comparar dos regiones para hacer un grafico
# lo voy a hacer una funcion

def comparacion(df, column1, column2, xlabel='', ylabel='', title=''):
    df_de_comparacion = df.groupby(by=[column1, column2]).agg({column1: ['count']})
    df_de_comparacion = df_de_comparacion.unstack()
    df_de_comparacion.columns = df_de_comparacion.columns.get_level_values(2)
    
    fig = plt.figure(figsize=[10,6])
    ax = fig.add_subplot(1,1,1)

    df_de_comparacion.plot.bar(ax=ax)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_title(title)
    plt.show()
    return df_de_comparacion
```

### Comparación de Stage con la Region

```python
region_stage = comparacion(df, 'Stage', 'Region', title='Comparación del Stage por Region', ylabel='Oportunidades')
```

### Comparación entre los porcentaje de Stage de cada Region

```python
normalized_region_stage = region_stage / region_stage.sum()
```

```python
fig = plt.figure(figsize=[10,6])
ax = fig.add_subplot(1,1,1)

normalized_region_stage.plot.bar(ax=ax)
ax.set_xlabel('')
ax.set_ylabel('Oportunidades normalizada [%]')
ax.set_title('Comparación normalizada del Stage por Region')
plt.show()
```

### Relación entre la cantidad de dinero y los productos en cada región

```python
region_stage_ASP = df.loc[:,['Region', 'Stage', 'ASP_(converted)']] #.set_index(['Region', 'Stage'])
```

```python
sns.distplot(region_stage_ASP['ASP_(converted)'])
```

```python
region_stage_ASP = region_stage_ASP.loc[region_stage_ASP['ASP_(converted)'] < 10]
region_stage_ASP.plot.box()
```

```python
region_stage_ASP = region_stage_ASP.loc[region_stage_ASP['ASP_(converted)'] < 0.75]
region_stage_ASP.plot.box()
```

```python
regiones = region_stage_ASP['Region'].cat.categories
estados = region_stage_ASP['Stage'].cat.categories
fig = plt.figure(figsize=[20,20])
    
for i, region in enumerate(regiones):
    tmp = region_stage_ASP.loc[region_stage_ASP['Region'] == region, ['Stage', 'ASP_(converted)']]
    ax = fig.add_subplot(3,2, i+1)
    for estado in estados:
        _tmp = tmp.loc[tmp['Stage'] == estado, 'ASP_(converted)']
        if (_tmp.size > 4):
            sns.distplot(_tmp, ax=ax)
    ax.legend(estados)
    ax.set_title(f'Comparación del ASP y el Stage en la region \'{region}\'')
    ax.set_xlabel('Precio promedio a la venta (USD)')
    ax.grid()

plt.show()
```

## Comparación con el codigo burocratico

#### Como se estan distribuyendo los codigos burocraticos en distintas regiones

```python
region_burocratic = comparacion(df, 'Region', 'Bureaucratic_Code')
```

```python
region_normalized_burocratic = (region_burocratic.T / region_burocratic.T.sum()).T
```

```python
fig = plt.figure(figsize=[10,6])
ax = fig.add_subplot(1,1,1)

region_normalized_burocratic.plot.bar(ax=ax)
ax.set_xlabel('')
ax.set_ylabel('Oportunidades normalizada [%]')
ax.set_title('Comparación normalizada de los codigos burocratricos usados Region')
plt.show()
```
