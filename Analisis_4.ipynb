```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
df['indice_convertidor']=df['ASP_(converted)']/df['ASP']
df.head()
```

```python
df['Total_Amount(converted)']=(df['Total_Amount']*df['indice_convertidor'])//1
df['indice_convertidor'].value_counts()
```

```python
df.head()
```

```python
df_closed=df.loc[(df['Stage']=='Closed Lost') | (df['Stage']=='Closed Won')]
df_closed['Stage'].value_counts()
```

```python
closed= df_closed.groupby(['Region','Delivery_Year','Stage'])\
.agg({'Opportunity_ID':['count'],'Total_Amount(converted)':['sum']}).dropna()
closed.head()
```

```python
level0 = closed.columns.get_level_values(0)
level1 = closed.columns.get_level_values(1)
closed.columns= level0 + '_' + level1
closed =closed.reset_index()
closed
```

```python
labels = ['2016', '2017', '2018', '2019']
Won = [313,622, 821, 37]
Lost = [399, 790, 883, 68]

x = np.arange(len(labels))  # the label locations
width = 0.2  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, Won, width, label='Closed Won')
rects2 = ax.bar(x + width/2, Lost, width, label='Closed Lost')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('# de Oportunidades')
ax.set_title('EvoluciÃ³n de oportunidades - RegiÃ³n: Americas')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```
