```python
labels = ['2016', '2017', '2018', '2019']
Won = [313,622, 821, 37]
Lost = [399, 790, 883, 68]

x = np.arange(len(labels))  # the label locations
width = 0.2  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, Won, width, label='Closed Won')
rects2 = ax.bar(x + width/2, Lost, width, label='Closed Lost')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('# de Oportunidades')
ax.set_title('EvoluciÃ³n de oportunidades - RegiÃ³n: Americas')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```

```python
import matplotlib
import squarify
```

```python
pie = df.loc[:,['Opportunity_ID','Total_Amount(converted)','Stage']]
##Podemos observar que tenemos valores nulos en Total_Amount
pie = pie.drop_duplicates(subset = ["Opportunity_ID"])
pie = pie.dropna(subset = ["Total_Amount(converted)"])
pie = pie.groupby('Stage').agg({'Total_Amount(converted)':'mean'})
pie=pie.reset_index()
pie = pie.sort_values(by="Total_Amount(converted)",ascending=False)
names = pie['Stage']
size = pie['Total_Amount(converted)']
pie.head()
```

```python
squarify.plot(sizes=size, label=names, color=["red","green","blue", "grey","yellow"],alpha=.5 )

fig = plt.gcf()
ax = fig.add_subplot()
fig.set_size_inches(16, 4.5)

plt.title("Promedio de Total_Amount(converted) segÃºn Stage",fontsize=20,fontweight="bold")
plt.axis('off')
plt.show()
```

```python
scatter = df.loc[:,['Opportunity_Name','Total_Amount(converted)','ASP_(converted)','Stage']]

scatter = scatter.drop_duplicates(subset = ["Opportunity_Name"])
scatter['Total_Amount(converted)']= scatter['Total_Amount(converted)'].fillna(0)
scatter['ASP_(converted)']= scatter['ASP_(converted)'].fillna(0)
```

```python
import seaborn as sns

g=sns.relplot(x="Total_Amount(converted)", y="ASP_(converted)",
              hue="Stage", style="Stage",
            kind="scatter", data=scatter).axes.flatten()
g[0].set_title("Relacion del estado entre ASP vs Monto total",fontsize=15)
g[0].set_xlabel("Monto Total (USD)",fontsize=10)
g[0].set_ylabel("Precio promedio de la venta (USD)",fontsize=10)
```
