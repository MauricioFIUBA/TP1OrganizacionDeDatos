```python
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
```

```python
closed.head()
```

```python
closed = closed.groupby(by=['Region', 'Delivery_Year', 'Stage']).agg({'Opportunity_ID_count': ['sum']}).dropna()
```

```python
closed = closed.reset_index()
closed.columns = closed.columns.get_level_values(0)
```

```python
# Si se quiere imprimir los 5 plots en una sola imagen
# Descomentar la lina que dice DESCOMENTAR
# Comantar la linea que dice COMENTAR
# Modificar el argumento de add_subplot por lo que dice en el comentario

regiones = closed['Region'].cat.categories

# fig = plt.figure(figsize=[10,10]) # DESCOMENTAR
for i, region in enumerate(regiones):
    fig = plt.figure(figsize=[10,10]) # COMENTAR
    tmp = closed.loc[closed['Region'] == region]
    tmp = tmp.pivot(index='Delivery_Year', columns='Stage', values='Opportunity_ID_count')
    ax = fig.add_subplot(1,1,1) # CAMBIAR POR (3, 2, i+1)
    tmp.plot.bar(ax=ax)
    for graficoDeBarras in ax.containers[:]:
        autolabel(graficoDeBarras)
    ax.set_title(f'Evolución de oportunidades - Región: {region}')
    ax.set_xlabel('')
    plt.xticks(rotation=0)
    ax.grid()

plt.show()
```

```python
import matplotlib
import squarify
```

```python
pie = df.loc[:,['Opportunity_ID','Total_Amount(converted)','Stage']]
##Podemos observar que tenemos valores nulos en Total_Amount
pie = pie.drop_duplicates(subset = ["Opportunity_ID"])
pie = pie.dropna(subset = ["Total_Amount(converted)"])
pie = pie.groupby('Stage').agg({'Total_Amount(converted)':'mean'})
pie=pie.reset_index()
pie = pie.sort_values(by="Total_Amount(converted)",ascending=False)
names = pie['Stage']
size = pie['Total_Amount(converted)']
pie.head()
```

```python
squarify.plot(sizes=size, label=names, color=["red","green","blue", "grey","yellow"],alpha=.5 )

fig = plt.gcf()
ax = fig.add_subplot()
fig.set_size_inches(16, 4.5)

plt.title("Promedio de Total_Amount(converted) según Stage",fontsize=20,fontweight="bold")
plt.axis('off')
plt.show()
```

```python
scatter = df.loc[:,['Opportunity_Name','Total_Amount(converted)','ASP_(converted)','Stage']]

scatter = scatter.drop_duplicates(subset = ["Opportunity_Name"])
scatter['Total_Amount(converted)']= scatter['Total_Amount(converted)'].fillna(0)
scatter['ASP_(converted)']= scatter['ASP_(converted)'].fillna(0)
```

```python
import seaborn as sns

g=sns.relplot(x="Total_Amount(converted)", y="ASP_(converted)",
              hue="Stage", style="Stage",
            kind="scatter", data=scatter).axes.flatten()
g[0].set_title("Relacion del estado entre ASP vs Monto total",fontsize=15)
g[0].set_xlabel("Monto Total (USD)",fontsize=10)
g[0].set_ylabel("Precio promedio de la venta (USD)",fontsize=10)
```
